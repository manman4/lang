```python
import pandas as pd
import matplotlib.pyplot as plt

# サンプルデータ
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 1, 8, 7]
}
df = pd.DataFrame(data)

# 散布図の作成
plt.figure(figsize=(8, 6))  # 図のサイズを指定
plt.scatter(df['x'], df['y'], color='blue', label='Data Points')  # 散布図
plt.title('Scatter Plot', fontsize=16)  # タイトル
plt.xlabel('X-axis', fontsize=14)  # x軸ラベル
plt.ylabel('Y-axis', fontsize=14)  # y軸ラベル
plt.legend()  # 凡例
plt.grid(True)  # グリッドを表示
plt.show()  # プロットを表示
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# サンプルデータ
data = {
    'x': np.random.uniform(0, 10, 50),  # 0から10の範囲でランダムなx値
    'y': np.random.uniform(0, 10, 50)  # 0から10の範囲でランダムなy値
}
df = pd.DataFrame(data)

# 散布図の作成
plt.figure(figsize=(8, 8))  # 図のサイズを指定
plt.scatter(df['x'], df['y'], color='blue', label='Data Points')  # 散布図

# グリッド線を追加
x_ticks = np.linspace(df['x'].min(), df['x'].max(), 4)  # x軸を3分割
y_ticks = np.linspace(df['y'].min(), df['y'].max(), 4)  # y軸を3分割
plt.xticks(x_ticks)  # x軸の目盛り設定
plt.yticks(y_ticks)  # y軸の目盛り設定
plt.grid(True, which='both', linestyle='--', color='gray', alpha=0.7)  # グリッド線

# タイトルとラベル
plt.title('Scatter Plot with 3x3 Grid', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# 凡例
plt.legend()

# プロット表示
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# サンプルデータ
data = {
    'x': np.random.uniform(0, 10, 50),  # 0から10の範囲でランダムなx値
    'y': np.random.uniform(0, 10, 50)  # 0から10の範囲でランダムなy値
}
df = pd.DataFrame(data)

# 分割する区間（自由に設定可能）
x_divisions = [2, 5]  # x軸の区分
y_divisions = [3, 6]  # y軸の区分

# 散布図の作成
plt.figure(figsize=(8, 8))
plt.scatter(df['x'], df['y'], color='blue', label='Data Points')

# グリッド線を追加
for x in x_divisions:
    plt.axvline(x=x, color='gray', linestyle='--', alpha=0.7)  # 縦線（x軸の区分）
for y in y_divisions:
    plt.axhline(y=y, color='gray', linestyle='--', alpha=0.7)  # 横線（y軸の区分）

# 軸範囲の設定（自動調整）
plt.xlim(0, 10)  # x軸の範囲
plt.ylim(0, 10)  # y軸の範囲

# タイトルとラベル
plt.title('Scatter Plot with Custom Grid Divisions', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# 凡例
plt.legend()

# プロット表示
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# サンプルデータ
np.random.seed(42)
data = {
    'x': np.random.uniform(0, 10, 50),  # 0から10の範囲でランダムなx値
    'y': np.random.uniform(0, 10, 50),  # 0から10の範囲でランダムなy値
    'group': np.random.choice(['A', 'B', 'C'], 50)  # ランダムにグループを割り当て
}
df = pd.DataFrame(data)

# 分割する区間（自由に設定可能）
x_divisions = [2, 5, 8]  # x軸の区分
y_divisions = [3, 6, 9]  # y軸の区分

# 散布図の作成
plt.figure(figsize=(8, 8))

# グループごとに色を設定
colors = {'A': 'red', 'B': 'blue', 'C': 'green'}

for group, group_data in df.groupby('group'):
    plt.scatter(group_data['x'], group_data['y'], 
                label=f'Group {group}', 
                color=colors[group], 
                alpha=0.7)

# グリッド線を追加
for x in x_divisions:
    plt.axvline(x=x, color='gray', linestyle='--', alpha=0.7)  # 縦線（x軸の区分）
for y in y_divisions:
    plt.axhline(y=y, color='gray', linestyle='--', alpha=0.7)  # 横線（y軸の区分）

# 軸範囲の設定（自動調整）
plt.xlim(0, 10)  # x軸の範囲
plt.ylim(0, 10)  # y軸の範囲

# タイトルとラベル
plt.title('Scatter Plot with Groups and Custom Grid Divisions', fontsize=16)
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# 凡例
plt.legend()

# プロット表示
plt.show()
```
