```python
import pandas as pd

# とある時系列が逆のデータを読み込む
df = pd.read_csv('./data.csv', parse_dates=['date'], index_col='date')
```

# 時系列が逆でもちゃんと図はおかしくない

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.xticks(rotation=90)
ax.plot(df.index, df['value01'])
ax.plot(df.index, df['value02'])
plt.show()
```

```python
df_random = df.sample(frac=1)
print(df_random)
```

# 時系列をランダムにすると、図はおかしくなる

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.xticks(rotation=90)
ax.plot(df_random.index, df_random['value01'])
ax.plot(df_random.index, df_random['value02'])
plt.show()
```
