```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 
np.random.seed(42)  # 
data = {
    'age': np.random.randint(18, 65, size=100),  # 
    'salary': np.random.normal(50000, 15000, size=100),  # 
    'score': np.random.randint(0, 100, size=100),  # 
}
df = pd.DataFrame(data)

# 
target_column = 'salary'

plt.figure(figsize=(8, 6))
sns.histplot(df[target_column], kde=True, bins=20, color='purple')  # 
plt.title(f'Distribution of {target_column}', fontsize=16)
plt.xlabel(target_column, fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
